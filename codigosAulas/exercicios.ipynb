```python
frutas = ['laranja', 'maçã', 'pera', 'banana', 'kiwi', 'maçã', 'banana']
print(frutas)

print("frutas.count('maçã')")
print(frutas.count('maçã'))

print("frutas.index('banana')")
print(frutas.index('banana'))

print("frutas.index('banana', 4)")
print(frutas.index('banana', 4))

print("frutas.reverse()")
frutas.reverse()
print(frutas)

print("frutas.append('uva')")
frutas.append('uva')
print(frutas)

print("frutas.sort()")
frutas.sort()
print(frutas)

print("frutas.pop()")
frutas.pop()
print(frutas)

frutas2 = ['melancia', 'abacaxi', 'cereja']
print("frutas2")
print(frutas2)

frutas.extend(frutas2)
print("frutas.extend(frutas2)")
print(frutas)

frutas.insert(0, 'uva')
print("frutas.insert(0, 'uva')")
print(frutas)

frutas.remove('maçã')
print("frutas.remove('maçã')")
print(frutas)

frutas.pop(1)
print("frutas.pop(1)")
print(frutas)

frutas.sort(key=len)
print("frutas.sort(key=len)")
print(frutas)

frutas.sort(key=len, reverse=True)
print("frutas.sort(key=len, reverse=True)")
print(frutas)

frutas3 = frutas.copy()
print("frutas3")
print(frutas3)

frutas3.clear()
print("frutas3.clear()")
print(frutas3)
```
